## NumPy Hello World

Let us start with the simplest possible NumPy examples.

NumPy is a library for working with arrays.  

An array is like a list, but it is more powerful for numerical computing.

```python
import numpy as np

# This is the standard import for NumPy.
# We will use the alias "np" throughout the notebook.
```

```python
# Create a simple NumPy array from characters.
# This is just a fun "hello world" example to show that arrays can hold strings too.

hello_world = np.array(["h", "e", "l", "l", "o", "w", "o", "r", "l", "d"])
print(hello_world)
```

```python
# Create an array of numbers from 0 to 24.
# np.arange(25) gives us 25 values: 0, 1, 2, ..., 24.

arr = np.arange(25)
print(arr)
```

```python
# Every NumPy array has a data type.
# The dtype tells us what kind of values are stored in the array.

print("dtype:", arr.dtype)
```

## Shape

The shape of an array tells us how many dimensions it has and how many items are in each dimension.

For example:
- a vector has shape like `(25,)`
- a matrix has shape like `(5, 5)`

```python
# Reshape the 1D array into a 5x5 matrix.
# This does not change the values, only the layout.

arr_2d = arr.reshape(5, 5)
print(arr_2d)
```

```python
# Check the shape of the original 1D array.

print("Original shape:", arr.shape)
```

```python
# Check the shape of the reshaped 2D array.

print("Reshaped array shape:", arr_2d.shape)
```

## Indexing

Indexing means selecting one item from an array.

Remember:
- NumPy uses zero-based indexing
- the first item is at index 0
- the second item is at index 1

```python
# Access a single value from the 1D array.
# This gets the value at index 8.

arr[8]
```

```python
# Get a slice of values from index 1 up to, but not including, index 5.
# This is a very common NumPy operation.

arr[1:5]
```

## Arrays in Two Dimensions

2D arrays are very important because they behave like tables or matrices.

You can think of them as:
- rows
- columns

```python
# Create a small 2D array manually.

arr_2d = np.array([
    [5, 10, 15],
    [20, 25, 30],
    [35, 40, 45]
])

print(arr_2d)
```

```python
# Select the second row.
# In NumPy, row indexing starts from 0.

arr_2d[1]
```

```python
# Get one individual element.
# Both forms below do the same thing.

print(arr_2d[1][0])
print(arr_2d[1, 0])
```

```python
# Slice a 2D array.
# This selects the top-right 2x2 block.

arr_2d[:2, 1:]
```

## Image-like Arrays

Images can also be stored as NumPy arrays.

A color image usually has shape:
- `(height, width, channels)`

For RGB images, the number of channels is usually 3:
- red
- green
- blue

```python
# Create a blank image-like array.
# This is a 64x64 image with 3 color channels.
# dtype=np.uint8 means pixel values are integers from 0 to 255.

img = np.zeros((64, 64, 3), dtype=np.uint8)

print("shape:", img.shape)
print("dtype:", img.dtype)
```

```python
# Extract the red channel.
# The result is a 2D array with shape (64, 64).

red = img[:, :, 0]
print(red.shape)
```

```python
# Crop the top-left 32x32 region.
# Slicing is a powerful way to extract part of an image.

crop = img[:32, :32]
print(crop.shape)
```

```python
# Flip the image horizontally.
# The ::-1 slice reverses the order along the width dimension.

flipped = img[:, ::-1]
print(flipped.shape)
```

## Arithmetic Operations

NumPy lets you do math directly on arrays.

This is one of the main reasons NumPy is so useful.

```python
# Create a small numeric array for arithmetic examples.

arr = np.arange(0, 10)
print(arr)
```

```python
# Element-wise addition.
# Each value is added to the matching value in the other array.

print(arr + arr)
```

```python
# Element-wise multiplication.

print(arr * arr)
```

```python
# Element-wise subtraction.

print(arr - arr)
```

```python
# Division by the array itself.
# The first value is 0 / 0, which becomes NaN.
# NumPy will usually show a warning, but it will still continue.

print(arr / arr)
```

```python
# Division by zero from the other direction.
# 1 / 0 becomes infinity.

print(1 / arr)
```

```python
# Power operation.
# This raises each element to the third power.

print(arr ** 3)
```

## Matrix Operations

Matrices are 2D arrays, and they are very common in machine learning.

In this section, we will practice:
- addition
- multiplication
- matrix multiplication
- transpose
- averages
- broadcasting
- distance calculations

```python
# Create two matrices.

A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

B = np.array([
    [7, 8, 9],
    [10, 11, 12]
])

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Shape of A:", A.shape)
```

```python
# Element-wise addition.
# This works because A and B have the same shape.

print("A + B:\n", A + B)
```

```python
# Element-wise multiplication.
# This is NOT matrix multiplication.
# It multiplies matching entries.

print("A * B:\n", A * B)
```

```python
# Scalar multiplication.
# Every element in A is multiplied by 2.

print("A * 2:\n", A * 2)
```

```python
# Matrix multiplication.
# The number of columns in A must match the number of rows in C.

C = np.array([
    [1, 2],
    [3, 4],
    [5, 6]
])

result = A @ C

print("A shape:", A.shape)
print("C shape:", C.shape)
print("A @ C:\n", result)
```

```python
# Transpose swaps rows and columns.

A_T = A.T

print("A:\n", A)
print("A transpose:\n", A_T)
```

```python
# Mean of all elements in the matrix.

print("Mean of all elements:", A.mean())
```

```python
# Mean across rows.
# axis=1 means "go across columns for each row".

print("Mean across rows:", A.mean(axis=1))
```

```python
# Mean across columns.
# axis=0 means "go down rows for each column".

print("Mean across columns:", A.mean(axis=0))
```

```python
# Sum across columns.

print("Sum across columns:", A.sum(axis=0))
```

```python
# Broadcasting example.
# A 1D vector can be added to each row of A automatically.

v = np.array([1, 1, 1])

result = A + v

print("Vector v:", v)
print("A + v:\n", result)
```

```python
# Broadcasting is one of the most important NumPy ideas.
# It lets us do fast vectorized operations without writing loops.

print("A shape:", A.shape)
print("v shape:", v.shape)
```

```python
# Euclidean distance between two vectors.
# This is a very important idea for nearest-centroid classification.

x = np.array([1.0, 2.0, 3.0])
y = np.array([4.0, 5.0, 6.0])

distance = np.linalg.norm(x - y)

print("Distance between x and y:", distance)
```
